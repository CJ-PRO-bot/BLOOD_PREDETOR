```python
#Assignment 2:
import random

def number_guessing_game():
    """
    This function plays a number guessing game with the user.
    The program picks a secret number, and the user tries to guess it.
    It keeps asking for a guess until the user is correct.
    """

    # 1. Picks a secret number between 1 and 10 (you can change the range)
    secret_number = random.randint(1, 10)

    # Initialize the user's guess to a value that won't match the secret number initially
    guess = None

    print("I've picked a secret number between 1 and 10. Can you guess it?")

    # 3. Keep asking until the user guesses correctly (Using a while loop)
    # The loop continues as long as the guess does not equal the secret number
    while guess != secret_number:
        try:
            # 2. Asks the user to guess the number
            guess = int(input("Enter your guess: "))

            # Optional: Provide hints to make the game more engaging
            if guess < secret_number:
                print("Too low! Try again.")
            elif guess > secret_number:
                print("Too high! Try again.")
        except ValueError:
            # Handle cases where the user enters something that's not a number
            print("That's not a valid number. Please enter an integer.")

    # 4. Print "Congratulations! You guessed correctly" when correct
    print(f"Congratulations! You guessed correctly. The number was {secret_number}.")

# Call the function to run the game
number_guessing_game()

#ASSIGNMENT 1:
def calculate_cinema_price():
    """
    Calculates and prints a cinema ticket price based on age, with a discount for Wednesday.
    """

    # 1. Asks the user to enter their age
    try:
        age = int(input("Please enter your age: "))
    except ValueError:
        print("Invalid input. Please enter a valid number for your age.")
        return # Exit the function if input is not a valid number

    # 2. Uses an if statement to decide the ticket price
    if age < 12:
        price = 300
    elif 12 <= age <= 17:
        price = 500
    elif 18 <= age <= 59:
        price = 800
    else:  # age >= 60
        price = 400

    # Extra challenge: Add a discount for Wednesday
    # You would need to import the 'datetime' module to check the day of the week
    import datetime
    today = datetime.datetime.now().strftime("%A") # Gets the full name of the day (e.g., "Wednesday")

    if today == "Wednesday":
        price -= 100
        print("Happy Wednesday! You received a 100 off discount.")

    # 3. Prints Your ticket price is xxx where xxx is the amount
    print(f"Your ticket price is {price}.")

# Call the function to run the program
calculate_cinema_price()

#ASSIGNMNET 3:
def describe_pet(animal_type, pet_name):
    """Prints a simple description of a pet."""
    print(f"I have a {animal_type}.")
    print(f"My {animal_type}'s name is {pet_name.title()}.")

# Call with positional arguments (order matters)
describe_pet('hamster', 'harry')

# Call with keyword arguments (order doesn't matter)
describe_pet(pet_name='harry', animal_type='hamster')

def sum_all(*numbers):
    """Calculates the sum of all numbers passed to the function."""
    total = 0
    for number in numbers:
        total += number
    return total

# Example usage
print(f"The sum of 1, 2, 3 is: {sum_all(1, 2, 3)}")
print(f"The sum of 10, 20, 30, 40 is: {sum_all(10, 20, 30, 40)}")

def print_user_info(**info):
    """Prints user information from a dictionary of keyword arguments."""
    print("User Info:")
    for key, value in info.items():
        print(f"  {key.title()}: {value}")

# Example usage
print_user_info(name='Alice', age=30, city='New York')
print_user_info(first_name='Bob', last_name='Smith')

def process_data(required_arg, *optional_pos_args, **optional_key_args):
    """Shows how to use all types of arguments together."""
    print(f"Required argument: {required_arg}")
    print(f"Optional positional arguments (args): {optional_pos_args}")
    print(f"Optional keyword arguments (kwargs): {optional_key_args}")

# Example usage
process_data(10, 20, 30, name='Charlie', role='Developer')

#ASSIGNMENT 4:
def calculate_grade(marks):
    """
    Calculates the grade based on a given mark (0-100).

    Args:
        marks (int): The student's score, expected to be between 0 and 100.

    Returns:
        str: The corresponding grade (A, B, C, D, or Failed).
    """

    # 1. Define the function with an if-elif-else statement to determine the grade
    if 90 <= marks <= 100:
        return "A"
    elif 80 <= marks < 90:
        return "B"
    elif 70 <= marks < 80:
        return "C"
    elif 60 <= marks < 70:
        return "D"
    elif 0 <= marks < 60:
        return "Failed"
    else:
        return "Invalid marks" # Handle cases outside the 0-100 range

def main():
    """
    Main function to get user input and display the calculated grade.
    """
    try:
        # 2. Ask the user to enter their marks
        user_marks = int(input("Enter your marks (0-100): "))

        # 3. Call the function and print the result
        grade = calculate_grade(user_marks)
        print(f"Your grade is: {grade}")

    except ValueError:
        print("Invalid input. Please enter a number between 0 and 100.")

# Run the main function
if __name__ == "__main__":
    main()
```
